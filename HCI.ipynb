```python
# Label Encoding
mapping_gender= {
    'F' : 0,
    'M' : 1
}

mapping_own_car = {
    'N' : 0,
    'Y' : 1
}

mapping_own_realty = {
    'N' : 0,
    'Y' : 1
}

mapping_education = {
    'Secondary / secondary special' : 0,
    'Higher - Academic Degree' : 1
}

mapping_emergency = {
    'No' : 0,
    'Yes' : 1
}

new_df2['CODE_GENDER'] = new_df2['CODE_GENDER'].map(mapping_gender)
new_df2['FLAG_OWN_CAR'] = new_df2['FLAG_OWN_CAR'].map(mapping_own_car)
new_df2['FLAG_OWN_REALTY'] = new_df2['FLAG_OWN_REALTY'].map(mapping_own_realty)
new_df2['NAME_EDUCATION_TYPE'] = new_df2['NAME_EDUCATION_TYPE'].map(mapping_education)
new_df2['EMERGENCYSTATE_MODE'] = new_df2['EMERGENCYSTATE_MODE'].map(mapping_emergency)
```

```python
# One Hot Encoder
# WARNING: just run this code 1 time
for cat in ['NAME_CONTRACT_TYPE', 'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE',
            'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE']:
    onehots = pd.get_dummies(new_df2[cat], prefix=cat)
    new_df2 = new_df2.join(onehots)
```

```python
# drop column after one hot encoding 
# WARNING: just run this code 1 time
new_df_train_2 = new_df2.drop(columns=['NAME_CONTRACT_TYPE', 'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'OCCUPATION_TYPE',
                                    'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 'FONDKAPREMONT_MODE', 'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE'])

new_df_train_2.info()
```

```python
# We change null values from CODE_GENDER with modus
new_df_train_2['CODE_GENDER'].fillna(new_df_train_2['CODE_GENDER'].mode()[0],inplace=True)

# We change null values from FLAG_OWN_CAR with modus
new_df_train_2['FLAG_OWN_CAR'].fillna(new_df_train_2['FLAG_OWN_CAR'].mode()[0],inplace=True)

# We change null values from FLAG_OWN_REALTY with modus
new_df_train_2['FLAG_OWN_REALTY'].fillna(new_df_train_2['FLAG_OWN_REALTY'].mode()[0],inplace=True)

# We change null values from NAME_EDUCATION_TYPE with modus
new_df_train_2['NAME_EDUCATION_TYPE'].fillna(new_df_train_2['NAME_EDUCATION_TYPE'].mode()[0],inplace=True)

# We change null values from EMERGENCYSTATE_MODE with modus
new_df_train_2['EMERGENCYSTATE_MODE'].fillna(new_df_train_2['EMERGENCYSTATE_MODE'].mode()[0],inplace=True)

new_df_train_2.info()
```

```python
# Models
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# Evaluation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import precision_recall_curve, plot_precision_recall_curve, average_precision_score
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import validation_curve
```

```python
from datetime import datetime as dt
from collections import defaultdict

def experiment(X_train,X_test,y_train,y_test):
    result = defaultdict(list)
    
    knn = KNeighborsClassifier()
    logreg = LogisticRegression()
    dtc = DecisionTreeClassifier()
    rf = RandomForestClassifier()
    grad = GradientBoostingClassifier()

    
    list_model = [('K-Nearest Neighbor',knn),
                  ('Logistic Regression',logreg),
                  ('Decision Tree',dtc),
                  ('Random Forest',rf),
                  ('Gradient Boosting',grad),
                 ]

    for model_name,model in list_model:
        start = dt.now()
        model.fit(X_train,y_train)
        duration = (dt.now()-start).total_seconds()
        
        y_pred = model.predict(X_test)
        
        accuracy = accuracy_score(y_test,y_pred)
        recall = recall_score(y_test,y_pred)
        precision = precision_score(y_test,y_pred)
        f_score = f1_score(y_test, y_pred)
        fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label=1) 
        roc_auc = auc(fpr, tpr)
        result['model_name'].append(model_name)
        result['model'].append(model)
        result['accuracy'].append(accuracy)
        result['recall'].append(recall)
        result['precision'].append(precision)
        result['f1_score'].append(f_score)
        result['roc_auc'].append(roc_auc)
        result['duration'].append(duration)
        
    return result
```

```python
result = experiment(X_train,X_test,y_train,y_test)
result = pd.DataFrame(result)
result
```
